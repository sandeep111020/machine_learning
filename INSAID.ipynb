```
 
import pandas as pd
from sklearn.model_selection import train_test_split
 
dataset_url_train = '/content/Churntrain.csv'
dataset_url_test = '/content/Churntest.csv'
data_train = pd.read_csv(dataset_url_train)
data_test = pd.read_csv(dataset_url_test)
data_test.head()
```

```
# droping unnescessary columns
data_train.drop(['customerID','PaymentMethod'],axis='columns',inplace=True) 
data_test.drop(['customerID','PaymentMethod'],axis='columns',inplace=True) 
data_test.head()
```

```
#Filtering

data_train= data_train.replace("Yes", 1)
data_train= data_train.replace("No", 0)
data_train= data_train.replace("No phone service", 0)
data_train= data_train.replace("No internet service", 0)
data_train= data_train.replace("Fiber optic", 1)
data_train= data_train.replace("DSL", 2)
data_train= data_train.replace("Month-to-month", 0)
data_train= data_train.replace("Two year",2)
data_train= data_train.replace("One year",1)

data_test= data_test.replace("Yes", 1)
data_test= data_test.replace("No", 0)
data_test= data_test.replace("No phone service", 0)
data_test= data_test.replace("No internet service", 0)
data_test= data_test.replace("Fiber optic", 1)
data_test= data_test.replace("DSL", 2)
data_test= data_test.replace("Month-to-month", 0)
data_test= data_test.replace("Two year",2)
data_test= data_test.replace("One year",1)

data_test.head()
```

```
dummies_gender_train = pd.get_dummies(data_train.gender)
dummies_gender_train.head()
```

```
# appending the dummies into file
data_train = pd.concat([data_train,dummies_gender_train],axis='columns')
data_train.drop(['gender'],axis='columns',inplace=True) 
data_train.head()
```

```
dummies_gender_test = pd.get_dummies(data_test.gender)
dummies_gender_test.head()
```

```
# merging dummies in file
data_test = pd.concat([data_test,dummies_gender_test],axis='columns')
data_test.drop(['gender'],axis='columns',inplace=True) 
data_test.head()
```

```
#  here we convert the string data into  float
for n,v in data_train.items():
    if v.dtype == "object":
        data_train[n] = v.factorize()[0]

for n,v in data_test.items():
    if v.dtype == "object":
        data_test[n] = v.factorize()[0]        
```

```
# separteing the data into test and train
X_train = data_train.drop(['Churn'],axis='columns')
X_test = data_test.drop(['Churn'],axis='columns')
y_train = data_train.Churn
y_test = data_test.Churn
```

```
X_train.info()
```

```
X_test.info()
```

```
# here we using random forest model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=200)


```

```
# here we train the data
model.fit(X_train,y_train)
```

```
# predicting the Churndata
y_test = model.predict(X_test)
y_test
```
